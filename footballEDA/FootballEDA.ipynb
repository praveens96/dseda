```python
######----with code reuse-------------##########
#this can be reused by function just replacing the column names AF, HF 
uniqueteamsfoulsdic = dict([(p,0) for p in bigdf["HomeTeam"].unique()])

def getTotalsofaMetric(onedf,uniqueteamsdic, homeCol, awayCol):
    for index, row in onedf.iterrows():  #loop over each year data
        uniqueteamsdic[row["HomeTeam"]] += row[homeCol] #add fouls count to unique teams dic of hometeam in a match
        uniqueteamsdic[row["AwayTeam"]] += row[awayCol] #add fouls count to unique teams dic of awayteam in a match
    allfoulscount = list(uniqueteamsdic.values()) #get fouls of each team and add. .values() return dict_values() type, so convert to list
    allfoulscount = pd.Series(allfoulscount).fillna(0).tolist() #to replace nan's converting to series and back to list.
    #allfoulscount.fillna(0)
    print(sum(i for i in allfoulscount)) #printing sum of indivual teams fouls. 

for onedf in dfslist:
    getTotalsofaMetric(onedf,uniqueteamsfoulsdic,"HF","AF")
print("Fouls From bigdf:"+str(bigdf["AF"].sum()+bigdf["HF"].sum()))  ##get sum from all matches for cross checking
print(uniqueteamsfoulsdic)
#####------------kicks-----------
uniqueteamskicksdic = dict([(p,0) for p in bigdf["HomeTeam"].unique()])
for onedf in dfslist:
    getTotalsofaMetric(onedf,uniqueteamskicksdic,"HC","AC")  ##HC-Home corner kics, AC-Away team corener kicks
print("Kicks From bigdf:"+str(bigdf["AC"].sum()+bigdf["HC"].sum()))  ##get sum from all matches for cross checking
print(uniqueteamskicksdic)
#####-----------shots---------
uniqueteamsshotsdic = dict([(p,0) for p in bigdf["HomeTeam"].unique()])
for onedf in dfslist:
    getTotalsofaMetric(onedf,uniqueteamsshotsdic,"HST","AST")  ##HC-Home corner kics, AC-Away team corener kicks
print("Shots From bigdf:"+str(bigdf["AST"].sum()+bigdf["HST"].sum()))  ##get sum from all matches for cross checking
print(uniqueteamsshotsdic)
#######-----red cards----------
uniqueteamsrcdic = dict([(p,0) for p in bigdf["HomeTeam"].unique()])
for onedf in dfslist:
    getTotalsofaMetric(onedf,uniqueteamsrcdic,"HR","AR")  ##HC-Home corner kics, AC-Away team corener kicks
print("Shots From bigdf:"+str(bigdf["AR"].sum()+bigdf["HR"].sum()))  ##get sum from all matches for cross checking
print(uniqueteamsrcdic)
#########--------------yellow cards----------
uniqueteamsycdic = dict([(p,0) for p in bigdf["HomeTeam"].unique()])
for onedf in dfslist:
    getTotalsofaMetric(onedf,uniqueteamsycdic,"HY","HY")  ##HC-Home corner kics, AC-Away team corener kicks
print("Yellow cards From bigdf:"+str(bigdf["AY"].sum()+bigdf["HY"].sum()))  ##get sum from all matches for cross checking
print(uniqueteamsycdic)
```
